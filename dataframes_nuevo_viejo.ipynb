```python
import pandas as pd
import numpy as np
```

```python
# Dataframes nuevo y viejo

df_new = pd.read_csv('df_BOGOTÁ.csv').sort_values('Nombre')
df_old = pd.read_csv('df_old.csv').sort_values('Nombre')
```

```python
# Productos de la validación manual

df_validacion_manual = pd.read_excel('validacion_manual.xlsx', names=['Productos', 'Cambios', 'Anterior', 'Categoria', 'EAN', 'Problema']).iloc[1:]
df_validacion_manual 
```

```python
# Filtramos el dataframe viejo con datos unicamente de Bogotá
# Cambiamos el formato de ean ya que hay datos numericos mezclados con letras (C100052, por ejemplo) que no permiten datos con
# formato tipo int.

df_old_bogota = df_old[df_old['Ciudad']=='Bogotá']
df_old_bogota.ean = df_old_bogota.ean.astype(str)
```

```python
# Listas con las columnas ean para el dataframe df_new y df_old_bogota

a = list(df_new.ean)
b = list(df_old_bogota.ean)
# Elementos nuevos que no están en el viejo
diferentes1 = list(elementos for elementos in a if elementos not in b)

# Elementos viejos que no están en el nuevo
diferentes2 = list(elementos for elementos in b if elementos not in a)

# Elementos nuevos que están tanto en el nuevo como en el viejo
iguales = list(elementos for elementos in b if elementos in a)
```

```python
# Dataframe con los datos nuevos que no estan en el viejo

df_diferentes1 = pd.DataFrame()
for i in range(0, len(diferentes1)):
    len_dato = len(df_new[df_new.ean==diferentes1[i]])
    if len_dato != 0:
        df1 = df_new[df_new.ean==diferentes1[i]]
        df_diferentes1 = pd.concat([df_diferentes1, df1])
```

```python
# Filtramos, ahora, los datos que no contienen precio

df_diferentes1_sin_precio = df_diferentes1[df_diferentes1['Precio Real'] == ' ' ]
```

```python
# Grabamos los dataframe resultantes

df_diferentes1.to_excel('df_diferentes1.xlsx', encoding='utf-8-sig', index=False)
df_diferentes1_sin_precio.to_excel('df_diferentes1_sin_precio.xlsx', encoding='utf-8-sig', index=False)
```
